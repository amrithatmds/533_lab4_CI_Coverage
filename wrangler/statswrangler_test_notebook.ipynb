```python
#from wrangler import statswrangler.outlierdrop as m1
import pandas as pd
import numpy as np
import wrangler.statswrangler.outlierdrop as mod1
import wrangler.statswrangler.training as mod2
```

```python
df = pd.read_csv("train.csv") #read example data train.csv
d = df[0:100] #use only 100 rows for easy visualization
d1 = d["SalePrice"]
```

```python
mod1.Data(d1,d).Outliers()
#returns the index of the outliers and the outliers itself as a tuple
```

```python
mod1.Dropped(d1,d).Outliers()
#Inheritance so the dropped object also has the Outliers function
```

```python
threshold = 0.1*len(d)
#suppose we say that we only want to drop no more than 10% of the data if they are outliers
dataframe = mod1.Dropped(d1,d,threshold).DropOutliers()

#The function DropOutliers removes the rows with outliers by index 
dataframe.head(20)
#Clearly we see that index 11 is now removed
dataframe[48:56] # we can see that 53 is removed as well
```

```python
train1, test1 = mod2.TrainingData(d).SplitTrain()
#This uses default 50:50 training data and test data
#returns a tuple (train, test)

train2, test2 = mod2.TrainingData(d, percentage = 0.8).SplitTrain()
#change percentage of training to 80% 

print(len(train1))
print(len(test1))

print(len(train2))
print(len(test2))

train1.head(3)
```
